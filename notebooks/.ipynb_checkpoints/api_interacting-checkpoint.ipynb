```python
q = "select * from coingecko_listed_projects where banned is false and narratives is null"
```

```python
df = pd.read_sql_query(q, con=connections.connect_postgres())
```

```python
df['checked'] = False
```

```python
df.head()
```

```python
no_webpage_in_cg = []
webpage_abandoned = []
connection_error = []
```

```python
print(webpage_abandoned[:])
```

```python
df = df[df['checked'] == False]
```

```python
for index, row in df.iloc[701:1009, :].iterrows():
    print(row["api_id"])
    time.sleep(8)
    data = dp.download_coingecko_market_info_for_api_id(row["api_id"])
    try:
        url = data['links']['homepage'][0]
        resp = requests.get(url, headers=agents.scrap())
        status_code = resp.status_code
        if status_code == 200:
            df.loc[index, 'checked'] = True
            continue
        else:
            print(f'Webpage abandoned. Status: {status_code}, page: {url}. | for {row["api_id"]}, index: {index}')
            webpage_abandoned.append(row['api_id'])
            df.loc[index, 'checked'] = True
    except KeyError:
        print(f'KeyError for {row["api_id"]}, index: {index}')
        print('\n')
        df.loc[index, 'checked'] = False
        continue
    except requests.exceptions.ConnectionError:
        print(f'Webpage ConnectionError - page: {url}. | for {row["api_id"]}, index: {index}')
        print('\n')
        connection_error.append(row['api_id'])
        df.loc[index, 'checked'] = False
    except requests.exceptions.MissingSchema:
        print(f'Webpage MissingSchema - for page: {url}. | for {row["api_id"]}, index: {index}')
        print('\n')
        no_webpage_in_cg.append(row['api_id'])
        df.loc[index, 'checked'] = False
```
